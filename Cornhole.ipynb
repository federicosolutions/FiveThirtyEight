## Imports

```python
import itertools
import math
from pandas import DataFrame
```

## Classes

```python
class Outcome:
    def __init__(self, label, probability, value):
        self.label = label
        self.probability = probability
        self.value = value
        
    def __repr__(self):
       return self.label
```

## Setup

```python
Values = [
    Outcome("Agressive Hole", .4, 3), 
    Outcome("Agressive Board", .3, 1),
    Outcome("Agressive Miss", .3, 0),
    Outcome("Conservative Hole", .1, 3),
    Outcome("Conservative Board", .8, 1),
    Outcome("Conservative Miss", .1, 0),
    Outcome("Wasted", 1, 0),
    ]
```

```python
print(Values)
```

```python
Permuntations = list(itertools.combinations_with_replacement(Values, 4))
print(len(Permuntations))
```

## Analysis

```python
df = DataFrame()
for p in Permuntations:
    temp = {}
    temp['Roll 1'] = p[0].label
    temp['Roll 2'] = p[1].label
    temp['Roll 3'] = p[2].label
    temp['Roll 4'] = p[3].label
    temp['Sum'] = int(p[0].value + p[1].value + p[2].value + p[3].value)
    temp['Raw Probability'] = (p[0].probability * p[1].probability * p[2].probability * p[3].probability)
    df = df.append(temp, ignore_index=True)

df
```

```python
p3 = df.loc[df['Sum'] == 3]
p3.sort_values(by='Raw Probability', ascending=False)
```
